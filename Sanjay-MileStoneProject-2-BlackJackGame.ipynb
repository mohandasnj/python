```python
import random

'''
    Player Account Class
'''
class PlayerAccount():
    '''
    Player initial account deposit balance
    Withdrawl method - for picking betting money
    Deposit method - for depositting bet money won
    '''
    def __init__(self,bankname,name,initialbal):
        self.bankname = bankname
        self.name = name
        self.balance = initialbal
        
    def deposit(self,deposit):
        self.balance += deposit
        return True
        
    def withdraw(self,withdraw):
        if (self.balance < withdraw):
            return False
        else:
            self.balance -= withdraw
            return True



'''
Create three classes namely
    Card
    Deck
    Player
'''
#print("Card, Deck and Player class")

'''
    Card class
        suit (ex: Spades, Diamonds, Clubs, Hearts)
        value (ex: Ac=1, .. Jack=11,Queen=12 & King=13)
        methods (init and show)
'''
class Card():
    def __init__(self,suit,value):
        self.suit = suit
        self.value = value
        
    def show(self):
        self_value = ""
        if (self.value == 1):
            self_value = "A"
        elif (self.value == 11):
            self_value = "J"
        elif (self.value == 12):
            self_value = "Q"
        elif (self.value == 13):
            self_value = "K"
        else:
            self_value = str(self.value)
    
        print("*------*")
        print(f"|{self_value.rjust(2,' ')}   {self.suit}|")
        #print("|      |")
        #print(f"|{self.suit}   {self_value.ljust(2,' ')}|")
        print("*------*")

'''
Deck class
        import random package to implement 
            import random” to use the Fisher Yates shuffle which is a very neat 
            shuffling algorithm that makes sure that each card has an equal likelihood
            of ending up in every other position.

        Deck class inherits Card class

        class variable suits list (ex: Spades, Clubs, Diamonds & Hearts)

        build method - build the deck 
            outer for loop iterates through suits
            inner for loop appends an instance of Card with v=1 to 13 for each suit

        init method - will initialize the deck of cards list and also calls build deck

        show method - iterates through cards list and shows each card by calling card's show method

        shuffle method - shuffles
            outer for loop goes from len of cards list -1 , all the way to 0 and strting at -1 index
            pick random number between index o and up until for loop curret index
            replace cards[i] with cards[r] and vice versa

        drawCard - mothod
            pops the latest card from cards list
'''
class Deck(Card):
    suits = ["S","C","D","H"]
    def __init__(self):
        self.cards=[]
        self.build()
    
    def build(self):
        for s in Deck.suits:
            for v in range(1,14):
                self.cards.append(Card(s,v))
    
    def show(self):
        for c in self.cards:
            c.show()
            
    def shuffle(self):
        for i in range(len(self.cards)-1,0,-1):
            r = random.randint(0,i)
            self.cards[i],self.cards[r] = self.cards[r],self.cards[i]
            
    def drawCard(self):
        return self.cards.pop()

'''
    init methood - initializes with name and an empty list of hand
    draw method - drwas a card from shuffled deck and appends to the hand list
    showHand method - itrates through hand ist and shows the card
'''
class Player():
    def __init__(self,name):
        self.name = name
        self.hand = []
        self.handtotal = 0
        
    def draw(self,deck):
        self.hand.append(deck.drawCard())
        return self
    
    def calculatetotal(self,ishigh="NO"):
        self.handtotal = 0
        for c in self.hand:
            if (c.value == 11 or c.value == 12 or c.value == 13):
                c.value = 10
            elif((c.value == 1) and (ishigh == "YES")):
                c.value = 11
            self.handtotal += c.value
        return self
                
    def showPlayerHand(self):
        for card in self.hand:
            card.show()
    
    def showDealerHand(self):
        for i, card in enumerate(self.hand):
            if i == 0:
                print("*------*")
                print(f"|**   X|")
                #print("|      |")
                #print(f"|X   **|")
                print("*------*")
            else:
                card.show()
'''
displayaccountdetails
'''
def displayaccountdetails(humanplayeraccount):
    print("Human Player Details :".title())
    print(f"Name                   : {humanplayeraccount.name}")
    print(f"Bank Name              : {humanplayeraccount.bankname}")
    print(f"Account Balance Amount : $ {humanplayeraccount.balance}")

'''
checkresult
'''
def playerbust(playerhand,ishigh="FALSE"):
    '''
        return NO_RESULT for no result
        "WINNER" for winner
        "LOSER" for loser
        "DRAW" for draw
    '''
    handtotal = 0
    for c in playerhand:
        if (c.value == 11 or c.value == 12 or c.value == 13):
            c.value = 10
        elif((c.value == 1) and (ishigh == "YES")):
            c.value = 11
        handtotal += c.value
    
    if (handtotal > 21):
        return "BUST"
    else:
        return "NO_RESULT"
        
def dealerbust(dealerhand,playertotal):
    '''
        return NO_RESULT for no result
        "WINNER" for winner
        "LOSER" for loser
        "DRAW" for draw
    '''
    handtotal = 0
    isvalueone = False
    for c in dealerhand:
        if (c.value == 11 or c.value == 12 or c.value == 13):
            c.value = 10
        elif(c.value == 1):
            c.value = 11
            isvalueone = True
        handtotal += c.value
    
    if(handtotal > 21 and isvalueone ):
        handtotal -= 10
    elif(handtotal < 21 and isvalueone):
        handtotal += 10
    
    if(playertotal == handtotal):
        return "DRAW"
    elif(handtotal > 21):
        return "BUST"
    elif(handtotal > playertotal):
        return "WINNER"
    else:
        return "NO_RESULT"
'''
Main program
'''   
if __name__ == "__main__":
    #print("Inside Main Program")
        
    human_player_name = input("Enter name of the human player : ")
    computer_dealer_name = "Computer Dealer"
    
    print("\nNow let create an account and deposit some dollars for the play\n")
    bankname = input("Enter name of the bank : ")
    
    print("\n")
    initial_deposit = int(input("Enter initial deposit dollar amount :"))
    print("\n")
    
    
    humanplayeraccount = PlayerAccount(bankname, human_player_name,initial_deposit)
    displayaccountdetails(humanplayeraccount)
    
    GAME_DONE = True
    
    while GAME_DONE:
        
        print(f"\nAccount Balance : ${humanplayeraccount.balance}")
        betamount = int(input("Enter bet amount : $"))
        
        if (humanplayeraccount.withdraw(betamount)):
            print(f"Succesfully withdrawn ${betamount} and your current account balance is ${humanplayeraccount.balance}")
        else:
            print(f"Not enough funds available to withdraw ${betamount} from your current account balance of ${humanplayeraccount.balance}")
            break
    
        deck = Deck()
        #deck.show()
        deck.shuffle()
        #deck.show()
        
        humanplayer = Player(humanplayeraccount.name)
        humanplayer.draw(deck)
        humanplayer.draw(deck)
        
        print("\nPlayer Hand : \n")
        humanplayer.showPlayerHand()
        
        computerdealer = Player(computer_dealer_name)
        computerdealer.draw(deck)
        computerdealer.draw(deck)
        
        print("\nDealer Hand : \n")
        computerdealer.showDealerHand()
        
        
        human_player_hit = True
        ishigh = "FALSE"

        while human_player_hit:
            if(playerbust(humanplayer.hand,ishigh) == "NO_RESULT"):
                stay = input("Do you want to hit or stay ? (hit/stay)").upper()
                if stay == "STAY":
                    human_player_hit = False
                    computer_dealer_hit = True
                else:
                    humanplayer.draw(deck)
                    if(ishigh == "FALSE"):
                        for c in humanplayer.hand:
                            if (c.value == 1):
                                ishigh = input("Do you want to replace the value of 'A' to be 11 ?  (yes/no) :").upper()
                                break
                    print("\nPlayer Hand : \n")
                    humanplayer.showPlayerHand()    
            elif(playerbust(humanplayer.hand,ishigh)=="BUST"):
                print(f"\nSorry {human_player_name} you lost the game !")
                print("\nPlayer Hand : \n")
                humanplayer.showPlayerHand()
                computer_dealer_hit = False
                break
          
        if (computer_dealer_hit):
            print("\n Dealer Plays \n")
            print("*************")
            humanplayer.calculatetotal(ishigh)
            #print(humanplayer.handtotal)
          
        while computer_dealer_hit:
            if(dealerbust(computerdealer.hand,humanplayer.handtotal) == "NO_RESULT"):
                    computerdealer.draw(deck)
            elif(dealerbust(computerdealer.hand,humanplayer.handtotal)=="WINNER"):
                print(f"\nCongratulations {computer_dealer_name} you won !")
                print("\nDealer Hand : \n")
                computerdealer.showPlayerHand()
                computer_dealer_hit = False
                break
            elif(dealerbust(computerdealer.hand,humanplayer.handtotal)=="BUST"):
                print(f"\nSorry {computer_dealer_name} you lost the game !")
                print("\nDealer Hand : \n")
                computerdealer.showPlayerHand()
                computer_dealer_hit = False
                betdeposited = humanplayeraccount.deposit(betamount*2)
                break
            elif(dealerbust(computerdealer.hand,humanplayer.handtotal)=="DRAW"):
                print(f"\nGame is draw !")
                print("\nDealer Hand : \n")
                computerdealer.showPlayerHand()
                computer_dealer_hit = False
                betdeposited = humanplayeraccount.deposit(betamount)
                break
        
        displayaccountdetails(humanplayeraccount)
        play_again = input("Do you want to play again ? (yes/no) : ").upper()
        if play_again == "NO":
            GAME_DONE = False
     
    print("\nThanks for playing BlackJack Game !\n\n\n\n\n\n")
```
