```python

symbol_player1 = ""
symbol_player2 = ""
readytoplay = False
current_position = 0
current_player = "Player 1"
first_player = ""
second_player = ""
current_symbol = ""
gameList = [" "]*9
replay = False

import os
clearScreen = lambda: os.system('clear')

#Print board position    
def printBoard(positionList,emptyBoardOnly):
    clearScreen()
    padding = "                      "
    padding_1 = "          >>          "
    if(emptyBoardOnly):
        print(padding + "Board Position")
        print(padding + "*-----------*")  
        print(padding + "| 1 | 2 | 3 |")
        print(padding + "+-----------+")  
        print(padding + "| 4 | 5 | 6 |")
        print(padding + "+-----------+")  
        print(padding + "| 7 | 8 | 9 |")
        print(padding + "*-----------*")
    else:
        print("\n")
        print("{0}'s Marker : {1}".format(first_player,symbol_player1) + \
            padding + "{0}'s Marker : {1}"\
            .format(second_player,symbol_player2))
        print("\n")
        print("Enter 'Exit' to exit the game at any time.")
        print("\n")
        print(padding + "Board Position" + padding + "Game Board")
        print(padding + "*-----------*" + padding + "*-----------*")
        print(padding + "| 1 | 2 | 3 |" + padding + "| {0} | {1} | {2} |"\
            .format(positionList[0],positionList[1],positionList[2]))
        print(padding + "*-----------*" + padding + "+-----------+")  
        print(padding + "| 4 | 5 | 6 |" + padding_1 + "| {0} | {1} | {2} |"\
            .format(positionList[3],positionList[4],positionList[5]))
        print(padding + "*-----------*" + padding + "+-----------+")  
        print(padding + "| 7 | 8 | 9 |" + padding + "| {0} | {1} | {2} |"\
            .format(positionList[6],positionList[7],positionList[8]))
        print(padding + "*-----------*" + padding + "*-----------*")

def winnerCheck(myList):
    all_xxx = "XXX"
    all_OOO = "OOO"
    if (myList[0] + myList[1] + myList[2] in (all_xxx,all_OOO)):
        return True
    elif (myList[3] + myList[4] + myList[5] in (all_xxx,all_OOO)):
        return True
    elif (myList[6] + myList[7] + myList[8] in (all_xxx,all_OOO)):
        return True
    elif (myList[0] + myList[3] + myList[6] in (all_xxx,all_OOO)):
        return True
    elif (myList[1] + myList[4] + myList[7] in (all_xxx,all_OOO)):
        return True
    elif (myList[2] + myList[5] + myList[8] in (all_xxx,all_OOO)):
        return True
    elif (myList[0] + myList[4] + myList[8] in (all_xxx,all_OOO)):
        return True
    elif (myList[2] + myList[4] + myList[6] in (all_xxx,all_OOO)):
        return True
    else:
        return False

print("\nWelcome to Tic Tac Toe!")
print("\n Use the below 'numpad' to match numbers to the grid on a tic \
    tac toe board.\n At any time enter 'Exit'to exit out of the game")

positionList = [" "]*9
printBoard(positionList,True)

readytoplayResponse = input("\nAre you ready to play ? Enter Yes or No : ").upper()

if readytoplayResponse == "YES":
    readytoplay = True


done = False
if (readytoplay):
    
    first_player = input("Enter first player name : ")
    second_player = input("Second player name : ")
    current_player = first_player

    symbol_player1 = input("{0}: Do you want to be X or O ? : ".format(first_player)).upper()

    if symbol_player1 == "X":
        symbol_player2 = "O"
    elif(symbol_player1 == "O"):
        symbol_player2 = "X"
    elif(symbol_player1 not in ("X","O")):
        done = True
        print("\nInvalid entry restart the game again\n\n")

    i = 1
    while (not done):
        if i==1 :
            print("\nNew Game started. This is {0}'s turn.".format(current_player))

        if (i==10):
            play_again = input("Game is draw, do you want to play again ? Yes or No : ").upper()
            if play_again == "YES":
                i=0
                gameList = [" "]*9
                printBoard(gameList,False)
            else:
                done = True
                print("Thanks for playing the game !")
                break
        else:
            if (current_symbol == ""):
                current_symbol = symbol_player1
                current_player = first_player
            
            current_position = input("\n{0} [ symbol {1} ] choose your next position (1 - 9) : "\
                .format(current_player,current_symbol))
            
            if(current_position.upper() == 'EXIT'):
                print("\nThanks for playing the game")
                print("\n\n\n")
                break
            
            try:
                val = int(current_position)
            except ValueError:
                print("No.. input is not a number. It's a string, replay")
                replay = True
            
            if ((val <= 9) and (gameList[val-1] == " ")):
                gameList[val-1] = current_symbol
                printBoard(gameList,False)
                
                if(winnerCheck(gameList)):
                    print("\nCongratulations ! {0} you have won the game.".format(current_player))
                    readytoreplay = input("\nAre you ready to play ? Enter Yes or No : ").upper()

                    if readytoreplay == "YES":
                        i=0
                        gameList = [" "]*9
                        printBoard(gameList,False)
                    else:
                        print("\nThanks for playing the game")
                        print("\n\n\n")
                        break
            elif ((val <= 9) and ((gameList[val-1] == "X") or (gameList[val-1] == "O"))):
                print("Cell is occupied, choose an empty space, replay")
                replay = True
            elif (val > 9):
                print("No.. input is not a number between 1 to 9 replay")
                replay = True
            
            if (replay):
                i -= 1
                replay = False
            else:
                if (current_symbol == "X"):
                    current_symbol = "O"
                elif (current_symbol == "O"):
                    current_symbol = "X"


                if(current_player == first_player):
                    current_player = second_player
                elif(current_player == second_player):
                    current_player = first_player
                    
            i += 1
    
else:
    print("\nThanks for playing the game\n\n\n\n\n\n\n\n\n\n")

    
```
